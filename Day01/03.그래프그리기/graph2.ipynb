### 2차원 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return (x - 2) * x * (x + 2)
print(f(1))
```

```python
print(f(np.array([1, 2, 3]))) 
# x는 ndarray 배열이며 각각에 대응한 f를 한꺼번에 ndarray로 돌려줌
```

```python
# 그래프를 그리는 x의 범위를 -3에서 3까지로 하고, 그 범위에서 계산한 x를 간격 0.5로 정의함.
x = np.arange(-3, 3.5, 0.5)
print(x)
```

```python
# linspace(n1, n2, n)하면 범위 n1에서 n2 사이를 일정 간격 n개의 구간으로 나눈 값을 돌려줌.
x = np.linspace(-3, 3, 10)
print(np.round(x, 2))
```

```python
plt.plot(x, f(x))
plt.show()
```

```python
# 두개의 인자값을 갖는 함수 
def f2(x, w):
    return (x - w) * x * (x + 2)

# x를 정의
x = np.linspace(-3, 3, 100) # x를 100분할

# 차트 show
plt.plot(x, f2(x, 2), color='black', label='$w=2$')
plt.plot(x, f2(x, 1), color='cornflowerblue', label='$w=1$')
plt.legend(loc="upper left")
plt.ylim(-15, 15)
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```
